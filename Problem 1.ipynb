```python
from scipy.stats import pearsonr
import numpy as np

datasetFile = open('./kc_house_data.csv', 'r')
price = []
bedrooms = []
bathrooms = []
sqft_living = []
sqft_lot = []
floors = []
waterfront = []
view = []
condition = []
grade = []
sqft_above = []
sqft_basement = []
yr_built = []
yr_renovated = []
zipcode = []
lat = []
long = []
sqft_living15 = []
sqft_lot15 = []
datasetFile.readline()

for line in datasetFile:
    splitedLine = line.split(',')
   
    price.append(float(splitedLine[2]))
    bedrooms.append(float(splitedLine[3]))
    bathrooms.append(float(splitedLine[4]))
    sqft_living.append(float(splitedLine[5]))
    sqft_lot.append(float(splitedLine[6]))
    a = splitedLine[7]
    floors.append(float(a[1:-1]))
    waterfront.append(float(splitedLine[8]))
    view.append(float(splitedLine[9]))
    condition.append(float(splitedLine[10]))
    grade.append(float(splitedLine[11]))
    sqft_above.append(float(splitedLine[12]))
    sqft_basement.append(float(splitedLine[13]))
    yr_built.append(float(splitedLine[14]))
    yr_renovated.append(float(splitedLine[15]))
    a = splitedLine[16]
    zipcode.append(float(a[1:-1]))
    a = splitedLine[17]
    lat.append(float(a[1:-1]))
    long.append(float(splitedLine[18]))
    sqft_living15.append(float(splitedLine[19]))
    sqft_lot15.append(float(splitedLine[20]))
    
#Bedrooms
average = float(sum(bedrooms))/len(bedrooms)
print("Bedrooms Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(bedrooms)), float(min(bedrooms)), np.var(bedrooms)))
print("Bedrooms Correlation Coefficient value with Price: %0.4f" % pearsonr(bedrooms, price)[0])
print()

#bathrooms
average = float(sum(bathrooms))/len(bathrooms)
print("Bathrooms Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(bathrooms)), float(min(bathrooms)), np.var(bathrooms)))
print("Bathrooms Correlation Coefficient value with Price: %0.4f" % pearsonr(bathrooms, price)[0])
print()

#sqft_living
average = float(sum(sqft_living))/len(sqft_living)
print("sqft_living Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_living)), float(min(sqft_living)), np.var(sqft_living)))
print("sqft_living Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_living, price)[0])
print()

#sqft_lot
average = float(sum(sqft_lot))/len(sqft_lot)
print("sqft_lot Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_lot)), float(min(sqft_lot)), np.var(sqft_lot)))
print("sqft_lot Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_lot, price)[0])
print()

#floors
average = float(sum(floors))/len(floors)
print("floors Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(floors)), float(min(floors)), np.var(floors)))
print("floors Correlation Coefficient value with Price: %0.4f" % pearsonr(floors, price)[0])
print()

#waterfront
average = float(sum(waterfront))/len(waterfront)
print("waterfront Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(waterfront)), float(min(waterfront)), np.var(waterfront)))
print("waterfront Correlation Coefficient value with Price: %0.4f" % pearsonr(waterfront, price)[0])
print()

#view
average = float(sum(view))/len(view)
print("view Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(view)), float(min(view)), np.var(view)))
print("view Correlation Coefficient value with Price: %0.4f" % pearsonr(view, price)[0])
print()

#condition
average = float(sum(condition))/len(condition)
print("condition Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(condition)), float(min(condition)), np.var(condition)))
print("condition Correlation Coefficient value with Price: %0.4f" % pearsonr(condition, price)[0])
print()

#grade
average = float(sum(grade))/len(grade)
print("grade Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(grade)), float(min(grade)), np.var(grade)))
print("grade Correlation Coefficient value with Price: %0.4f" % pearsonr(grade, price)[0])
print()

#sqft_above
average = float(sum(sqft_above))/len(sqft_above)
print("sqft_above Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_above)), float(min(sqft_above)), np.var(sqft_above)))
print("sqft_above Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_above, price)[0])
print()

#sqft_basement
average = float(sum(sqft_basement))/len(sqft_basement)
print("sqft_basement Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_basement)), float(min(sqft_basement)), np.var(sqft_basement)))
print("sqft_basement Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_basement, price)[0])
print()

#yr_built
average = float(sum(yr_built))/len(yr_built)
print("yr_built Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(yr_built)), float(min(yr_built)), np.var(yr_built)))
print("yr_built Correlation Coefficient value with Price: %0.4f" % pearsonr(yr_built, price)[0])
print()

#yr_renovated
average = float(sum(yr_renovated))/len(yr_renovated)
print("yr_renovated Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(yr_renovated)), float(min(yr_renovated)), np.var(yr_renovated)))
print("yr_renovated Correlation Coefficient value with Price: %0.4f" % pearsonr(yr_renovated, price)[0])
print()

#zipcode
average = float(sum(zipcode))/len(zipcode)
print("zipcode Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(zipcode)), float(min(zipcode)), np.var(zipcode)))
print("zipcode Correlation Coefficient value with Price: %0.4f" % pearsonr(zipcode, price)[0])
print()

#lat
average = float(sum(lat))/len(lat)
print("lat Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(lat)), float(min(lat)), np.var(lat)))
print("lat Correlation Coefficient value with Price: %0.4f" % pearsonr(lat, price)[0])
print()

#long
average = float(sum(long))/len(long)
print("long Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(long)), float(min(long)), np.var(long)))
print("long Correlation Coefficient value with Price: %0.4f" % pearsonr(long, price)[0])
print()

#sqft_living15
average = float(sum(sqft_living15))/len(sqft_living15)
print("sqft_living15 Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_living15)), float(min(sqft_living15)), np.var(sqft_living15)))
print("sqft_living15 Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_living15, price)[0])
print()

#sqft_lot15
average = float(sum(sqft_lot15))/len(sqft_lot15)
print("sqft_lot15 Average: %.2f, Max: %.2f, Min: %.2f Variance: %.2f" % (average, float(max(sqft_lot15)), float(min(sqft_lot15)), np.var(sqft_lot15)))
print("sqft_lot15 Correlation Coefficient value with Price: %0.4f" % pearsonr(sqft_lot15, price)[0])
print()
```
